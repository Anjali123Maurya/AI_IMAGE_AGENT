```python
# Step 1: Upload your image
from google.colab import files
uploaded = files.upload()

# Step 2: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Step 3: Load and preprocess image
filename = list(uploaded.keys())[0]
img = load_img(filename, target_size=(224, 224))
img_array = img_to_array(img) / 255.0

# Step 4: Analyze image
model = MobileNetV2(weights='imagenet')
img_processed = preprocess_input(np.expand_dims(img_array, axis=0))
predictions = model.predict(img_processed)
decoded = decode_predictions(predictions, top=3)[0]

# Step 5: Show predictions
print("\n AI Insights from the Image:")
for i, (imagenetID, label, prob) in enumerate(decoded):
    print(f"{i+1}. {label}: {prob * 100:.2f}% confidence")

# Step 6: Dominant color
image_rgb = (img_array * 255).astype(np.uint8)
pixels = image_rgb.reshape((-1, 3))
unique, counts = np.unique(pixels, axis=0, return_counts=True)
dominant = unique[np.argmax(counts)]
dominant_clean = tuple(int(v) for v in dominant)
print(f"\n Dominant Color (RGB): {dominant_clean}")

```
